```python
from langchain_classic.agents import AgentExecutor, create_react_agent
from langchain_classic import hub
```

```python
prompt = hub.pull("hwchase17/react")
print(prompt)
```

```python
@tool
def get_weather_data(city: str) -> str:
    """
    This function fetches the current weather data for a given city 
    """
    url = f"https://api.weatherstack.com/current?access_key=83ac90656e3f7643fa28b2e2c09cefe0&query=lucknow"

    response = requests.get(url)

    return response.json()
```

```python
agent = create_react_agent(
    llm=llm,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
# response = agent_executor.invoke({"input": "3 ways to reach Goa from Delhi"})
response = agent_executor.invoke({"input": "find the capital of Uttar Pradesh, then find it's current weather condition"})
print(response)
```
